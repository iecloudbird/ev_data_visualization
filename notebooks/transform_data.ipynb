# Change data formatting to a wide table

```python
import pandas as pd

df_iea = pd.read_csv('../data/raw/IEA Global EV Data 2024.csv')
df_iea.head()
```

```python
df_wide = df_iea.pivot_table(
    index=['region', 'year', 'category', 'mode', 'powertrain'],
    columns='parameter',
    values='value',
    aggfunc='first'
).reset_index()

df_wide.columns = [col.lower().replace(' ', '_') for col in df_wide.columns]
df_wide.head()
```

```python
df_wide.to_csv('../data/processed/iea_wide_format.csv', index=False)
print(f"Saved: {len(df_wide)} rows, {len(df_wide.columns)} columns")
```
