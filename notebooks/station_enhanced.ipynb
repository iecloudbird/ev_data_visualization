```python
import pandas as pd

df_stations = pd.read_csv('../data/raw/detailed_ev_charging_stations.csv')
```

```python
df_stations['city'] = df_stations['Address'].str.split(',').str[-1].str.strip()
```

```python
city_to_country = {
    'San Francisco': 'United States', 'Los Angeles': 'United States', 'Chicago': 'United States',
    'Bangkok': 'Thailand', 'Toronto': 'Canada', 'Mumbai': 'India',
    'São Paulo': 'Brazil', 'Moscow': 'Russia', 'Beijing': 'China',
    'Shanghai': 'China', 'Paris': 'France', 'Berlin': 'Germany',
    'London': 'United Kingdom', 'Tokyo': 'Japan', 'Seoul': 'Korea',
    'Sydney': 'Australia', 'Melbourne': 'Australia', 'Delhi': 'India'
}

df_stations['country'] = df_stations['city'].map(city_to_country)
```

```python
df_stations['is_fast_charger'] = df_stations['Charger Type'] == 'DC Fast Charger'
df_stations['is_renewable'] = df_stations['Renewable Energy Source'] == 'Yes'
df_stations['is_24_7'] = df_stations['Availability'] == '24/7'
df_stations['cost_per_full_charge'] = df_stations['Cost (USD/kWh)'] * 60
df_stations['station_age'] = 2024 - df_stations['Installation Year']

df_stations.head()
```

```python
df_stations.to_csv('../data/processed/stations_enhanced.csv', index=False)
print(f"Saved: {len(df_stations)} rows, {len(df_stations.columns)} columns")
```

```python
mapped = df_stations['country'].notna().sum()
total = len(df_stations)
print(f"Mapped {mapped}/{total} stations ({mapped/total*100:.1f}%)")
print(f"{total-mapped} stations have unknown country")
```
