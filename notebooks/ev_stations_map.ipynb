```python
import pandas as pd
import folium

df = pd.read_csv('../data/raw/detailed_ev_charging_stations.csv')

map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=2)

for idx, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Address']
    ).add_to(m)

m
```

```python
# Clean version show only stations on land

from global_land_mask import globe

df_land = df[df.apply(lambda row: globe.is_land(row['Latitude'], row['Longitude']), axis=1)].copy()

map_center = [df_land['Latitude'].mean(), df_land['Longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=2)

for idx, row in df_land.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=row['Address']
    ).add_to(m)

m
```
